```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import sem
import requests
import json
from pprint import pprint
```

```python
cities_df = pd.read_csv("worldcities.csv")
sample_df = cities_df.sample(500)
worldcities = sample_df.round(2)
worldcities.reset_index(inplace = True)
del worldcities['index']
worldcities.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = f"{url}&units=imperial&appid=86e99f8f52b689a868c3db258b66be67&q="
```

```python
cities = worldcities['City']
humidity = []
cityid = []
maxt = []
cloud = []
wind = []
date = []
country = []
lat = []
lon = []

for city in cities:
    
    response = requests.get(query_url + city)
    response2 = response.json()
    lat.append(response2['coord']['lat'])
    lon.append(response2['coord']['lon'])
    humidity.append(response2['main']['humidity'])
    cityid.append(response2['id'])
    maxt.append(response2['main']['temp_max'])
    cloud.append(response2['clouds']['all'])
    wind.append(response2['wind']['speed'])
    date.append(response2['dt'])
    country.append(response2['sys']['country'])
    
print(f"Retrieving Results...")

#try:
    #print(f"The city of {cities} has the ID {cityid}.")
        
#except (KeyError, IndexError):
    #print("Missing field/result... skipping.")
        
    #print("------------")
    
#print(f"The latitude information received is: {lat}")
#print(f"The longitude information received is: {lon}")
#print(f"The humidity information received is: {humidity}")
#print(f"The ID information received is: {cityid}")
#print(f"The max temperature information received is: {maxt}")
#print(f"The cloudiness information received is: {cloud}")
#print(f"The wind information received is: {wind}")
#print(f"The date information received is: {date}")
#print(f"The country information received is: {country}")
```

```python
weather_dict = {
    "City": cities,
    "Latitude": lat,
    "Longitude": lon,
    "Max temperature(F)": maxt,
    "Humidity(%)": humidity,
    "Cloudiness": cloud,
    "Wind Speed": wind,
    "Country": country,
    "Date": date,
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```
